```python
dataset = pdf2text.get_dataset()
```

```python
corpora = corpus.makeCleanCorpus(dataset, lemmatize=True, removePunct=True, removeNums=True,makeSentences=True, removeURL=True)
```

```python
corpora.keys()
```

```python
corpora["Ethereum"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
final = []
for k in corpora.keys():
    for i in corpora[k]:
        final += [i]
len(final)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(final)
```

```python
true_k = 10
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```

```python
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(" %s" % terms[ind])
```

```python
print("\n")
print("Prediction")
X = vectorizer.transform([corpora["Bitcoin"][10]])
predicted = model.predict(X)
print(predicted)
```
